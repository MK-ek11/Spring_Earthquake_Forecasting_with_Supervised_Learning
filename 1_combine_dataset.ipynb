# Combine all the dataset from 1900 till 2022

```python
## Combine the Dataset 
data_df_2001 = pd.read_csv("dataset/japan 1900-2001.csv")
data_df_2018 = pd.read_csv("dataset/Japan earthquakes 2001 - 2018.csv")
data_df_2022 = pd.read_csv("dataset/Japan 2018-2022.csv")
```

```python
#vertical_stack = pd.concat([data_df_2018, data_df_2001], axis=0)
vertical_stack = pd.concat([data_df_2018, data_df_2001], axis=0)
vertical_stack = pd.concat([data_df_2022, vertical_stack], axis=0)
```
